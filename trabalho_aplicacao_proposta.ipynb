### Metodologia

Cada cromossomo do GA será um conjunto de modelos, a priori de uma mesma família de algoritmos, como: Decision Trees ou Redes Neurais. Os Genes serão os modelos e os alelos serão os hiperparâmetros (ou caso sejam NNs, os alelos podem ser os pesos das redes). O número de genes do cromossomo será definido a partir do número de classes que existam no dataset, de forma que cada modelo terá uma classe específica como maior peso. Para cada geração os modelos classificarão o dataset de teste e a classe target será escolhida por votação. O fitness será a acurácia que for obtida por cada um dos indivíduos, de forma que quanto maior for a acurácia melhor é o indivíduo. A condição de parada será a estagnação do melhor indivíduo.
